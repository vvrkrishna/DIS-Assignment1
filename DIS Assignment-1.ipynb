```python
class program:
    n = 5
    PrintPattern(n)
    print('\r')

    n2 = 6
    PrintSeries(n2)
    print('\r')

    s = "09:15:35PM"
    t = USFTime(s)
    print(t)
    print('\r')

    n3 = 110
    k = 11
    USFnumbers(n3, k)
    print('\r')

    words = ["abcd", "dcba", "lls", "s", "sssll"]
    PalindromePairs(words)
    print('\r')

    n4=10
    Stones(10)
    
#This method prints pattern based on the input number using recursion.
    def PrintPattern(n): 
        try:
            k=0
            #If n is positive the number in the sequence decreases.
            if n>0:
                for j in range(0,n): 
                    #Printing the number with an additional space in the end
                    print(n-k, end=" ")
                    k = k + 1
                print("\r")
                if n>1:
                #Calling function to repeat the printing pattern decreasing value of n by '1' in each loop until n is 0.
                    PrintPattern(n-1)
                else:
                    pass
            #If n is negative the number in the sequence increases
            elif n<0:
                for j in range(n,0): 
                    print(n+k, end=" ")
                    k = k + 1
                #Move to the new line after printing the sequence
                print("\r")
                if n<1:
                #Calling function to repeat the printing pattern increasing value of n by '1' in each loop until n is 0.
                    PrintPattern(n+1)
                else:
                    pass
            #If n is zero sequence cannot be formed
            elif n==0:
                print('Series cannot be formed with value Zero')
        #Handling Exception
        except:
            print("Exception occured while computing PrintPattern")


    def PrintSeries(n):
        try:
            if n>0:
                #The loop continues from 1 till the number given in the input
                for k in range(1,n+1):
                #Logic behind each number in the sequence
                    f=(k*(k+1))/2
                    if k<n:
                        print(int(f),end=",")
                    else:
                        print(int(f))
            else:
                print('The minimum value given to n for series to be printed is 1')
        #Handling Exception
        except:
            print("Exception occured while computing PrintSeries")
            

    def USFTime(time):
        try:
            USF_time=''
            earth_time_seconds=None
            x=int(time[0:2]) #Extracting Hours
            y=int(time[3:5]) #Extracting Minutes
            z=int(time[6:8]) #Extracting Seconds
            if ((len(time)==10)&(x<=12)&(y<=59)&(z<=59)): #Validating the given input
                
                if (time[-2:]=='AM'): #Validating time format
                    
                    earth_time_seconds=(x*60*60)+(y*60)+(z) #Converting given time to earth seconds
                    
                    #Converting seconds in earth to USF time
                    USF_time=str(round(earth_time_seconds/2700))+":"+str(round((earth_time_seconds%2700)//45))+":"+str(round(earth_time_seconds%2700)%45)
                    return USF_time
                
                #Adding 12 hrs of equivalent time if earth time is in PM
                elif (time[-2:]=='PM'): #Validating time format
                    
                    earth_time_seconds=(x*60*60)+(y*60)+z+(43200) #Converting given time to earth seconds
                    
                    #Converting Seconds in earth to USF time
                    USF_time=str(round(earth_time_seconds/2700))+":"+str(round((earth_time_seconds%2700)//45))+":"+str(round(earth_time_seconds%2700)%45)
                    return USF_time
                
                else:
                    print('Invalid time format')
                    
            else:
                print('Please enter time in proper format')
                
        except:
            print("Exception occured while computing USFTime")
            
            
    def USFnumbers(t,m):
        try:
            def output(i):
            #Returning values according to thier divisibility with '3','5' & '7'
                if i%3==0 and i%5==0 and i%7==0:
                    return 'USF'
                elif i%3==0 and i%5==0 and i%7!=0:
                    return 'US'
                elif i%3!=0 and i%5==0 and i%7==0:
                    return 'SF'
                elif i%3==0 and i%5!=0 and i%7==0:
                    return 'UF'
                elif i%3==0 and i%5!=0 and i%7!=0:
                    return 'U'
                elif i%3!=0 and i%5==0 and i%7!=0:
                    return 'S'
                elif i%3!=0 and i%5!=0 and i%7==0:
                    return 'F'
                else:
                    return i
            for k in range(1,t+1):
                if k<t:
                    print(output(k),end=" ")
                else:
                    print(output(k))
                #Moving to new line until the remainder becomes 0
                if k%m==0:
                    print("\r") 
      #Handling Exception
        except:
            print("Exception occured while computing USFNumbers")
    
    
    def PalindromePairs(wordslist):
        try:
            list=[]
            for i in range(len(wordslist)):
                for j in range(i,len(wordslist)):
                    k=str(wordslist[i])+str(wordslist[j]) #Merging values in a list to get possible combinations
                    m=str(wordslist[j])+str(wordslist[i])
            #Checking if the string is same even after reversing it,if it is then appending to final list
                    if k==k[::-1] and i!=j:
                        list.append([i,j])
                    if m==m[::-1] and i!=j:
                        list.append([j,i])
            if len(list)==0:
                print('Palindromes are not found from the given set of words') #Prints this statement if none of palindromes are founds
            else:
                return list
        except:
            print("Exception occured while computing PalindromePairs")
    
    def Stones(n):
        try:
            numbers=[]
            possible_moves=[1,2,3]
            if isinstance(n,int)==True:
                if n>0 :
                    if n%4==0: #If input value is a multiple of 4 then there is no chance of winning.
                        print('False')
                    else:
                        #If number of stones are less than 4 then main player can easily win.
                        if 0<n<4:
                            numbers.append(n)
                            print(numbers)
                        else:
                            #The below function returns all the possible moves for the player to win the game.
                            def possible_comb(valuelist,total,index,sub_list):
                                final_moves=[[2,2],[3,1],[1,3]] #Final moves when the remaining stones are 4
                                #If we get exact total
                                if total==0:
                                    if len(sub_list)%2!=0:
                                        for k in range(len(final_moves)):
                                            print(sub_list+final_moves[k])
                                        return
                                #If the current total becomes negative
                                if total<0:
                                    return
                                for i in range(index,len(valuelist)):
                                    sub_list.append(valuelist[i])
                                    #Recursion from next numbers
                                    possible_comb(valuelist,total-valuelist[i],i,sub_list)
                                    #Removing number from list-Back tracking
                                    sub_list.remove(valuelist[i]) 
                            possible_comb(possible_moves,n-4,0,[])
            else:
                print('Entered value must be an integer greater than zero')
        except:
            print('Exception occured while computing Stones')
            
#If the input number is multiple of 4,using the given set of values the last move can never be the main player.
#For all other values we will only consider the cases where total number of moves is an odd number,since the final move should be of main player.
#In the above case the main player should only take certain moves when the remaining number of stones are 4(i.e, [1,3],[2,2],[3,1])
#I subracted 4 from the main number to consider all the possible cases.
```
